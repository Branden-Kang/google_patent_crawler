```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Branden-Kang/google_patent_crawler/main/data/logistics_data.csv',header=1)
df.head()
```

```python
df.shape
```

```python
# Patent Abstract Crawler
from urllib.request import urlopen
from bs4 import BeautifulSoup
from tqdm.auto import tqdm, trange
from time import sleep

for i in trange(len(df)):
    html = urlopen(df["result link"][i])
    # Web crawler
    bs = BeautifulSoup(html, "html.parser")

    # Extract abstract information using result link
    try:
        patent_abstract = bs.head.find("meta", {"name":"description"}).get('content')
    except AttributeError:
        continue
    else:
        # Add 'abstract' column
        df.loc[i, 'abstract'] = patent_abstract
    sleep(0.01)
```
